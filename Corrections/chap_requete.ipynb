````python
# requete imbriquée 

"""
pas de score pour cette partie de la requête 
{"term": {"play_name.keyword": "Hamlet"}}
{"range": {"year": {"gte": 1601}}}


ici Elasticsearch calcule un score 
{"match": {"text_entry": "love"}}
{"match": {"text_entry": "time"}}

"""

query = {
    "query": {
        "bool": {
            "must": [
                {"term": {"play_name.keyword": "Hamlet"}},
                {"range": {"year": {"gte": 1601}}},
                {
                    "bool": {
                        "should": [
                            {"match": {"text_entry": "love"}},
                            {"match": {"text_entry": "time"}}
                        ],
                        "minimum_should_match": 1
                    }
                }
            ]
        }
    }
}

"""
(play_name = Hamlet) AND (year >= 1601) AND (match love OU match time) 
"""

res = es.search(index="shakespeare", body=query)

for hit in res["hits"]["hits"]:
    print(hit["_score"], hit["_source"])
````

La différence concrète entre `0` et `1` apparaît ici :

* avec `minimum_should_match: 0`, on voit aussi :

```python
2.049822 {'play_name': 'Hamlet', 'speaker': 'HAMLET', 'text_entry': 'To be or not to be', 'year': 1603}
```

* avec `minimum_should_match: 1`, cette ligne disparaît.

## Ce que cela prouve

Ce document :

```json
{
  "play_name": "Hamlet",
  "speaker": "HAMLET",
  "text_entry": "To be or not to be",
  "year": 1603
}
```

respecte bien :

* `play_name = Hamlet`
* `year >= 1601`

mais il ne matche ni :

* `"love"`
* ni `"time"`

Donc :

### Avec `minimum_should_match = 0`

il est accepté, parce que les `should` sont optionnels.

### Avec `minimum_should_match = 1`

il est rejeté, parce qu’il faut au moins un match sur `love` ou `time`.

### Cas `0`

```text
(A AND B) + bonus éventuel si C ou D
```

### Cas `1`

```text
(A AND B) AND (C OR D)
```

## Pourquoi le reste de la liste ne change presque pas

Parce que les autres documents affichés contiennent déjà `love` ou `time`, donc ils satisfont aussi bien :

* la version souple (`0`)
* que la version stricte (`1`)

Leur score reste donc du même ordre, et ils restent dans les résultats.

```text
minimum_should_match = 0
→ les clauses should restent optionnelles

minimum_should_match = 1
→ au moins une clause should doit matcher
```


```text
ce document sort par exemple
{
  "speaker": "HAMLET",
  "play_name": "Hamlet",
  "year": 1605,
  "text_entry": "love is everywhere"
}
```

Oui, mais ce document sort dans les deux cas, puisque lui matche déjà `"love"`.

Le document le plus intéressant pour montrer la différence est justement :

```json
{
  "play_name": "Hamlet",
  "speaker": "HAMLET",
  "text_entry": "To be or not to be",
  "year": 1603
}
```


### Trouver les phrases contenant "love"

````python
query = {
    "query": {
        "match": {
            "text_entry": "love"
        }
    }
}

res = es.search(index="shakespeare", body=query)

for hit in res["hits"]["hits"]:
    print(hit["_source"])
````

### Trouver les phrases contenant "love" prononcées par Othello

````python
query = {
    "query": {
        "bool": {
            "must": [
                {"match": {"text_entry": "love"}},
                {"term": {"speaker.keyword": "OTHELLO"}}
            ]
        }
    }
}

res = es.search(index="shakespeare", body=query)

for hit in res["hits"]["hits"]:
    print(hit["_source"])
````

### Trouver toutes les phrases sauf celles de Macbeth

````python
query = {
    "query": {
        "bool": {
            "must_not": [
                {"term": {"play_name.keyword": "Macbeth"}}
            ]
        }
    }
}

res = es.search(index="shakespeare", body=query)

for hit in res["hits"]["hits"]:
    print(hit["_source"])
````

### Trouver les phrases contenant "be" et prononcées par Hamlet ou Othello

````python
# filter sera sans impact sur le score 
query = {
    "query": {
        "bool": {
            "must": [
                {"match": {"text_entry": "be"}}
            ],
            "filter": [
                {
                    "terms": {
                        "speaker.keyword": ["HAMLET", "OTHELLO"]
                    }
                }
            ]
        }
    }
}

res = es.search(index="shakespeare", body=query)

for hit in res["hits"]["hits"]:
    print(hit["_score"], hit["_source"])
````

````python
# should aura un impact sur le score 

res = es.search(index="shakespeare", body={
  "query": {
    "bool": {
      "must": [
                {"match": {"text_entry": "be"}},
            ],
      "should": [
        { "term": { "speaker.keyword": "HAMLET" } },
        { "term": { "speaker.keyword": "OTHELLO" } }
      ],
      "minimum_should_match": 1
    }
  }
})

for hit in res["hits"]["hits"]:
    print(hit["_score"], hit["_source"])
````

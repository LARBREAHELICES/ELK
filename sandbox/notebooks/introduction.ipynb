```python
import requests
import json

url = "http://elasticsearch:9200/shakespeare/_search"

response = requests.get(url)

data = response.json()
print(json.dumps(data, indent=2))
```

```python
url = lambda id : f"http://elasticsearch:9200/shakespeare/_doc/{id}"

docs = [
    {
    "play_name": "Hamlet",
    "speaker": "HAMLET",
    "text_entry": "To be, or not to be: that is the question."
    },
    {
      "play_name": "Macbeth",
      "speaker": "LADY MACBETH",
      "text_entry": "Out, damned spot! out, I say!"
    },
    {
      "play_name": "Othello",
      "speaker": "OTHELLO",
      "text_entry": "She loved me for the dangers I had passed."
    }
]

for i, doc in enumerate(docs):
    response = requests.put(url(i+1), json=doc)
    print(response.json())
```

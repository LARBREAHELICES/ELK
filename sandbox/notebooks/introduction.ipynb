```python


ES_URL = "http://elasticsearch:9200"

def run_search(query):
    resp = requests.get(f"{ES_URL}/shakespeare/_search", json=query)
    data = resp.json()
    print(json.dumps(data, indent=2, ensure_ascii=False))
    return data
```

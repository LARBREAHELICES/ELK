```python
# Chapitre 1
from elasticsearch import Elasticsearch, helpers

bulk = helpers.bulk

import json

es = Elasticsearch("http://elasticsearch:9200")
# print(es.info())
```

```python
res = es.search(
     index="shakespeare",
     query={ "term": {
      "speaker.keyword": "HAMLET"
    }}
)

#print(json.dumps(res["hits"]["hits"], indent=2))
```

```python
res = es.search(
     index="shakespeare",
     query={
         "term": {
          "speaker.keyword": {
          "value": "hamlet",
          "case_insensitive": True
      }
    }
})

#print(json.dumps(res["hits"]["hits"], indent=2))
```

```python
res = es.search(
     index="shakespeare",
     query={
         "bool": {
              "must": [
                { "match": { "text_entry": "be" } },
                { "term": { "speaker.keyword": "HAMLET" } }
              ]
        }
})
#print(json.dumps(res["hits"]["hits"], indent=2))
```

```python
res = es.search(
     index="shakespeare",
     query={
        "bool": {
            "should": [
                { "term": { "speaker.keyword": "HAMLET" } },
                { "term": { "speaker.keyword": "OTHELLO" } }
          ],
          "minimum_should_match": 1
    }
})
# print(json.dumps(res["hits"]["hits"], indent=2))
```

```python
res = es.search(
     index="shakespeare",
     query={
        "bool": {
            "should": [
                { "term": { "speaker.keyword": "HAMLET" } },
                { "term": { "speaker.keyword": "OTHELLO" } }
          ],
          "minimum_should_match": 0
    }
})
#print(json.dumps(res["hits"]["hits"], indent=2))
```

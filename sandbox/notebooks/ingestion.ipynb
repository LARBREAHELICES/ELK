```python
# ingestion des données 

actions = [
    {
        "_index": "shakespeare",
        "_source": doc
    }
    for doc in docs
]


if es.indices.exists(index="shakespeare"):
    es.indices.delete(index="shakespeare")
else:
    es.indices.create(index="shakespeare")

success, errors = bulk(es, actions, refresh=True)

print("Documents indexés :", success)
print("Erreurs :", errors)
```
